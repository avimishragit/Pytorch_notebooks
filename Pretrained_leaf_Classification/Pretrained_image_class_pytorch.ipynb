```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
from torch import nn
from torch.optim import Adam
import torchvision.transforms as transforms
from torchvision import models
from torch.utils.data import Dataset, DataLoader

from sklearn.preprocessing import LabelEncoder

from PIL import Image


```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
train = pd.read_csv('bean-leaf-lesions-classification/train.csv')
val = pd.read_csv('bean-leaf-lesions-classification/val.csv') 
display(train.head(),val.head())
print(train.shape,val.shape)
```

```python
data_df = pd.concat([train,val], ignore_index = True)
display(data_df.tail())
print(data_df.shape)
```

```python
image_ploting = 16
n_rows = n_cols = int(np.ceil(np.sqrt(image_ploting)))
fig, ax = plt.subplots(n_rows,n_cols)
for row in range(n_rows):
    for col in range(n_cols):
        image = data_df.sample(n=1)['image:FILE']
        ax[row,col].imshow(image)
        ax[row,col].axis('off')

plt.show()
```

```python
train = data_df.sample(frac = 0.7, random_state = 42)
test = data_df.drop(train.index)
```

```python
label_encoder = LabelEncoder()
transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor(),
    transforms.ConvertImageDtype(torch.float),
])
```

```python
class ImageData(Dataset):
    def __init__(self,dataframe, transform = None):
        self.dataframe = dataframe
        self.transform = transform
        self.labels = torch.tensor(label_encoder.fit_transform(dataframe['category'])).to(device)

    def __len__(self):
        return self.dataframe.shape[0]

    def __getitem__(self,index):
        image_path = self.dataframe.iloc[index,0]
        label = self.labels[index]
        image = Image.open(image_path).convert('RGB')
        if self.transform:
            image = (self.transform(image)/255).to(device)

        return image, label
        
```

```python
train_dataset = ImageData(dataframe = train,transform = transform)
test_dataset = ImageData(dataframe = test,transform = transform)
```

```python
image_ploting = 16
n_rows = n_cols = int(np.ceil(np.sqrt(image_ploting)))
f, ax = plt.subplots(n_rows, n_cols)
for row in range(n_rows):
    for col in range(n_cols):
        image = train_dataset[np.random.randint(0,train_dataset.__len__())][0].cpu()
        ax[row, col].imshow((image*255).squeeze().permute(1,2,0))
        ax[row, col].axis('off')

plt.tight_layout()
plt.show()
```

```python
LR = 1e-3
BATCH_SIZE = 32
EPOCHS = 2

train_loader = DataLoader(train_dataset, batch_size = BATCH_SIZE, shuffle = True)
test_loader = DataLoader(test_dataset, batch_size = BATCH_SIZE, shuffle = True)
```

```python
num_classes = len(data_df['category'].unique())
googlenet_model.fc = torch.nn.Linear(googlenet_model.fc.in_features, num_classes)# converting lastlayer out_features to number of categories
googlenet_model.to(device)
```

```python
from tqdm import tqdm
criterion = nn.CrossEntropyLoss()
optimizer = Adam(googlenet_model.parameters(),lr =LR)
total_loss_train_plot = []
total_acc_train_plot = []

for epoch in tqdm(range(EPOCHS),desc="Outer Loop",bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} [{percentage:.0f}%]'):
    total_acc_train = 0
    total_loss_train = 0

    for (inputs,labels) in tqdm(train_loader, desc="Inner Loop", leave=True):
        optimizer.zero_grad()
        outputs = googlenet_model(inputs)
        train_loss = criterion(outputs, labels)
        total_loss_train += train_loss.item()
        train_loss.backward()

        train_acc = (torch.argmax(outputs, axis = 1)==labels).sum().item()
        total_acc_train += train_acc
        optimizer.step()
    tqdm.write(f"Processed epoch {epoch}")
    total_loss_train_plot.append(round(total_loss_train/1000,4))
    total_acc_train_plot.append(round(total_acc_train/(train_dataset.__len__())*100,4))

    print(f'Epoch {epoch+1}/{EPOCHS}, \
        Train Loss: {round(total_loss_train/100, 4)} \
        Train Accuracy {round((total_acc_train)/train_dataset.__len__() * 100, 4)}%')
    print()
```

```python
with torch.no_grad():
    total_loss_test = 0
    total_acc_test = 0
    for index,(input, labels) in enumerate(test_loader):
        prediction = googlenet_model(input)
        acc = (torch.argmax(prediction,axis = 1)== labels).sum().item()
        total_acc_test += acc

print(f"Accuracy Score is: {round((total_acc_test/test_dataset.__len__())*100, 2)}%")
  
```

```python


fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

axs[0].plot(total_loss_train_plot, label='Training Loss')
axs[0].set_title('Training Loss over Epochs')
axs[0].set_xlabel('Epochs')
axs[0].set_ylabel('Loss')
axs[1].set_ylim([0, 2])
axs[0].legend()

axs[1].plot(total_acc_train_plot, label='Training Accuracy')
axs[1].set_title('Training Accuracy over Epochs')
axs[1].set_xlabel('Epochs')
axs[1].set_ylabel('Accuracy')
axs[1].set_ylim([0, 100])
axs[1].legend()

plt.tight_layout()

plt.show()
```
