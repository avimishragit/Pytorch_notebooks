```python
class ImageDataset(Dataset):
    def __init__(self,dataframe,transform = None):
        self.dataframe = dataframe
        self.transform = transform
        self.labels = torch.tensor(
            label_encoder.transform(dataframe['labels'])
        ).long().to(device)

    def __len__(self):
        return self.dataframe.shape[0]
    
    def __getitem__(self,index):
        image_path = self.dataframe.iloc[index,0]
        label = self.labels[index]
        image = Image.open(image_path).convert('RGB')
        if self.transform:
            image = self.transform(image).to(device)
    
        return image, label
```

```python
import torch
import torch.nn as nn

class CovNet(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)  # First Convolution layer
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)  # Second Convolution layer
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)  # Third Convolution layer
        self.pooling = nn.MaxPool2d(2, 2)  # The pooling layer
        self.relu = nn.ReLU()  # ReLU Activation function
        self.dropout = nn.Dropout(p=0.2)  # Dropout layer
        self.batch_norm1 = nn.BatchNorm2d(32)
        self.batch_norm2 = nn.BatchNorm2d(64)
        self.batch_norm3 = nn.BatchNorm2d(128)
        
        self.flatten = nn.Flatten()  # Flatten and vectorize the output feature maps
        self.linear1 = nn.Linear((128 * 16 * 16), 256)  # First Linear Layer
        self.linear2 = nn.Linear(256, 128)  # Second Linear Layer
        self.output = nn.Linear(128, len(data_df['labels'].unique()))  # Output Linear Layer

    def forward(self, x):
        x = self.conv1(x)
        x = self.batch_norm1(x)
        x = self.relu(x)
        x = self.pooling(x)
        x = self.dropout(x)

        x = self.conv2(x)
        x = self.batch_norm2(x)
        x = self.relu(x)
        x = self.pooling(x)
        x = self.dropout(x)

        x = self.conv3(x)
        x = self.batch_norm3(x)
        x = self.relu(x)
        x = self.pooling(x)
        x = self.dropout(x)

        x = self.flatten(x)
        x = self.linear1(x)
        x = self.relu(x)
        x = self.dropout(x)

        x = self.linear2(x)
        x = self.relu(x)
        x = self.dropout(x)

        x = self.output(x)
        return x
```

```python
for epoch in range(EPOCHS):
    total_acc_train = 0
    total_loss_train = 0
    total_loss_val = 0
    total_acc_val = 0

    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
    
        train_loss = criterion(outputs,labels)
        total_loss_train += train_loss.item()
        train_loss.backward()

        train_acc = (torch.argmax(outputs,axis =1)==labels).sum().item()
        total_acc_train += train_acc
        optimizer.step()

    with torch.no_grad():
        for inputs, labels in val_loader:
            outputs = model(inputs)
            val_loss = criterion(outputs, labels)
            total_loss_val += val_loss.item()

            val_acc = (torch.argmax(outputs,axis = 1)==labels).sum().item()
            total_acc_val += val_acc
 
    total_loss_train_plot.append(round(total_loss_train/1000, 4))
    total_loss_validation_plot.append(round(total_loss_val/1000, 4))
    total_acc_train_plot.append(round(total_acc_train/(train_dataset.__len__())*100, 4))
    total_acc_validation_plot.append(round(total_acc_val/(val_dataset.__len__())*100, 4))

    print(f'''Epoch {epoch+1}/{EPOCHS}, Train Loss: {round(total_loss_train/100, 4)} Train Accuracy {round((total_acc_train)/train_dataset.__len__() * 100, 4)}
              Validation Loss: {round(total_loss_val/100, 4)} Validation Accuracy: {round((total_acc_val)/val_dataset.__len__() * 100, 4)}''')
    print("="*50)

torch.save(model.state_dict(), 'convnet_model.pth')
```

```python
# Make predictions with the loaded model
def make_predictions(model, data_loader):
    """
    Make predictions using a loaded model on a given data loader.

    Parameters:
    - model: PyTorch model, the loaded model.
    - data_loader: PyTorch DataLoader, the data loader for predictions.
    """
    predictions = []
    with torch.no_grad():
        for inputs, labels in data_loader:
            outputs = model(inputs)
            batch_predictions = torch.argmax(outputs, axis=1)
            predictions.extend(batch_predictions.cpu().numpy())
    return predictions

# Get predictions on the validation set
predictions = make_predictions(loaded_model, val_loader)
print(f"Predictions: {predictions}")
```

```python
# dataset.py
import torch
from torch.utils.data import Dataset
from torchvision import transforms
from PIL import Image
import numpy as np
from sklearn.preprocessing import LabelEncoder

class ImageDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform
        self.labels = torch.tensor(
            LabelEncoder().fit_transform(dataframe['labels'])
        ).long()

    def __len__(self):
        return self.dataframe.shape[0]
    
    def __getitem__(self, index):
        image_path = self.dataframe.iloc[index, 0]
        label = self.labels[index]
        image = Image.open(image_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
    
        return image, label
```

```python
# utils.py
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score

def train(model, device, loader, criterion, optimizer):
    model.train()
    total_loss = 0
    correct = 0
    for batch_idx, (data, target) in enumerate(loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        _, predicted = torch.max(output.data, 1)
        correct += (predicted == target).sum().item()

    accuracy = correct / len(loader.dataset)
    return total_loss / len(loader), accuracy

def evaluate(model, device, loader, criterion):
    model.eval()
    total_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            loss = criterion(output, target)
            total_loss += loss.item()
            _, predicted = torch.max(output.data, 1)
            correct += (predicted == target).sum().item()

    accuracy = correct / len(loader.dataset)
    return total_loss / len(loader), accuracy
```
